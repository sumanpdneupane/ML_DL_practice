```python
f = open('iphone_sales_dataset.csv')
data = f.read()
f.close()
```

```python
datas = data.split('\n')
datas = datas[1:-1]
print(len(datas))
```

```python
for idx, data in enumerate(datas):
    print(f"idx: {idx+1} data: {data}")
```

```python
with open('iphone_sales_dataset.csv', 'r') as csvfile:
    reader = csvfile.read()
    data = reader.split('\n')
    data = data[1:-1]
    print(len(data))
    for idx, data in enumerate(data):
        print(f"idx: {idx+1} data: {data}")
```
