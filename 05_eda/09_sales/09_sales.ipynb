```python
data['Manager'] = data['Manager'].str.strip().replace(r'\s+', ' ', regex = True)
```

```python
data['Quantity'] = data['Quantity'].round()
```

```python
data.head(3)
```

```python
data['Quantity'] = data['Quantity'].astype(int)
```

```python
data.head(3)
```

```python
data.info()
```

```python
data['Payment Method'].value_counts().plot(kind='bar')
```

```python
data['Payment Method'].value_counts(normalize=True).plot(kind='bar')
```

```python
data.head(3)
```

```python
most_quantity = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
most_quantity= most_quantity.reset_index()
```

```python
plt.figure(figsize = (9,3))
plt.bar(most_quantity['Product'], most_quantity['Quantity'], width = 0.4, color = ['red', 'blue', 'green', 'yellow', 'cyan'])
plt.title('Most Selling Products - By Quantity')
plt.xlabel('Product')
plt.ylabel('Quantity')
plt.show()
```

```python
data.head(3)
```

```python
data['Revenue'] = data['Price'] * data['Quantity']
```

```python
data.head(3)
```

```python
most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
most_revenue
```

```python
most_revenue = most_revenue.reset_index()
most_revenue
```

```python
plt.figure(figsize=(8,5))
plt.bar(most_revenue['Product'], most_revenue['Revenue'], width = 0.4, color = ['red', 'blue', 'green', 'yellow', 'cyan'])
plt.title('Most Selling Products - By Revenue')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```

```python
data.head(3)
```

```python
city_revenue = data.groupby('City')['Revenue'].sum().sort_values(ascending=False)
city_revenue = city_revenue.reset_index()
city_revenue
```

```python
plt.figure(figsize=(8,5))
plt.bar(city_revenue['City'], city_revenue['Revenue'], width = 0.4, color = ['red', 'blue', 'green', 'yellow', 'cyan'])
plt.title('Maximum Revenue - By City')
plt.xlabel('City')
plt.ylabel('Revenue')
plt.show()
```
